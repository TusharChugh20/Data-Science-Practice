```python
import pandas as pd
```

```python
dataset=pd.read_csv("weather.csv")
```

```python
dataset
```

```python
# Return reshaped DataFrame organized by given index / column values.
dataset.pivot(index="date" ,columns="city")
```

```python
# columns: used to make new dataframe column
#index: used to make new dataframe rows
dataset.pivot(index="date" ,columns="city" ,values="humidity")
```

```python

dataset.pivot(index="city" ,columns="date" ,values="humidity")
```

```python
# Unpivot a DataFrame from wide to long format
dataset.melt(id_vars=['date', 'city'])
```

```python
dataset.melt(id_vars=['date', 'city'],value_vars=['temperature'])
```
