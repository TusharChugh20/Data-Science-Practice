```python
X=ds.data  #feature
# X=ds.data[:, np.newaxis, 2] #ask sir
print(X.shape)
y=ds.target  #value depending on feature
print(y.shape)
```

```python
# Splitting datset into 80% training datset and 20 % test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
                                        
```

```python
#oject of linear regression builtin model
lin_reg=linear_model.LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
predicted = lin_reg.predict(X_test)
```

```python
print(mean_squared_error(y_test, predicted))
```
