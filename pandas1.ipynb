```python
import pandas as pd
```

```python
# creation of dataframe
# A DataFrame is a table.
# Each entry corresponds to a row (or record) and a column.
pd.DataFrame({'Yes': [10, 20], 'No': [11, 21]})
```

```python
pd.DataFrame({'Tushar': ['I didnt like it.', 'It was awful.'], 'Shekhar': ['Pretty good.', 'Great Taste.']})
```

```python
#creation of series
#A Series is a sequence of data values.
pd.Series([1, 2, 3, 4, 5])
```

```python
#assigning row labels
pd.DataFrame({'Tushar': ['I liked it.', 'It was awful.'], 
              'Shekhar': ['Pretty good.', 'Great Taste.']},
             index=['Product A', 'Product B'])
```
